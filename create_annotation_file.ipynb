```python
import os
```

```python
def create_test_annot(folder, proliv):

    annot_list = []

    for img_name in os.listdir(folder):
        if proliv:
            annot_list.append(
                img_name + ' ' + str(1) + "\n"
            )
        else:
            annot_list.append(
                img_name + ' ' + str(0) + "\n"
            )

    return annot_list
```

```python
proliv_folder = "./homework_2_dataset/dataset/proliv/"
ne_proliv_folder = "./homework_2_dataset/dataset/train/"
```

```python
proliv_ann = create_test_annot(os.path.join(proliv_folder, "imgs"), True)
ne_proliv_ann = create_test_annot(os.path.join(ne_proliv_folder, "imgs"), False)
```
